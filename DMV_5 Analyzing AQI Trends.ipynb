```python
# Cleaning up name changes
data.state = data.state.replace({'Uttaranchal':'Uttarakhand'}) 
data.state[data.location == "Jamshedpur"] = data.state[data.location == 'Jamshedpur'].replace({"Bihar":"Jharkhand"})
```

```python
# Changing types to uniform format
types = {
    "Residential": "R", 
    "Residential and others": "RO",
    "Residential, Rural and other Areas": "RRO", 
    "Industrial Area": "I",
    "Industrial Areas": "I", 
    "Industrial": "I", 
    "Sensitive Area": "S", 
    "Sensitive Areas": "S", 
    "Sensitive": "S", 
    np.nan: "RRO"
}

data.type = data.type.replace(types) 
data.head()
```

```python
# defining columns of importance, which shall be used reguarly 
VALUE_COLS = ['so2', 'no2', 'rspm', 'spm', 'pm2_5']
```

```python
# invoking SimpleImputer to fill missing values
imputer = SimpleImputer(missing_values=np.nan, strategy='mean') 
data[VALUE_COLS] = imputer.fit_transform(data[VALUE_COLS])
```

```python
# checking to see if the dataset has any null values left over and the format 
print(data.isnull().sum())
data.tail()
```

```python
# Plotting highest and lowest ranking states
# defining a function to find and plot the top 10 and bottom 10 states for a given indicator (defaults to SO2) 
def top_and_bottom_10_states(indicator="so2"):
    fig, ax = plt.subplots(2,1, figsize=(20, 12))
    ind = data[[indicator, 'state']].groupby('state', as_index=False).median().sort_values(by=indicator,ascending=False)
    top10 = sns.barplot(x='state', y=indicator, data=ind[:10], ax=ax[0], color='red') 
    top10.set_title("Top 10 states by {} (1991-2016)".format(indicator)) 
    top10.set_ylabel("so2 (Âµg/m3)")
    top10.set_xlabel("State")
    bottom10 = sns.barplot(x='state', y=indicator, data=ind[-10:], ax=ax[1], color='green') 
    bottom10.set_title("Bottom 10 states by {} (1991-2016)".format(indicator)) 
    bottom10.set_ylabel("so2 (Âµg/m3)")
    bottom10.set_xlabel("State") 

top_and_bottom_10_states("so2") 
top_and_bottom_10_states("no2")
```

```python
def highest_levels_recorded(indicator="so2"): 
    plt.figure(figsize=(20,10))
    
    # Convert indicator column to numeric (ignore errors to handle bad data)
    data[indicator] = pd.to_numeric(data[indicator], errors='coerce')
    
    # Drop NaN values
    df = data[['state', indicator]].dropna(subset=[indicator])
    
    # Find the highest recorded level per state
    ind = df.groupby('state', as_index=False)[indicator].max()
    
    # Plot
    sns.barplot(x='state', y=indicator, data=ind)
    plt.title(f"Highest ever {indicator.upper()} levels recorded by state")
    plt.xticks(rotation=90)
    plt.ylabel(f"{indicator.upper()} level")
    plt.xlabel("State")
    plt.tight_layout()
    plt.show()
```

```python
highest_levels_recorded("so2")
highest_levels_recorded("no2")
highest_levels_recorded("rspm")
```

```python
import matplotlib.pyplot as plt

def type_avg(indicator=""):
    # Select only numeric columns for averaging
    numeric_cols = data[VALUE_COLS].select_dtypes(include='number').columns

    # Group by type and take mean only for numeric columns
    type_avg_df = data[numeric_cols.tolist() + ['type']].groupby("type").mean(numeric_only=True)

    # If an indicator is provided, plot only that column
    if indicator:
        type_avg_df[indicator].plot(kind='bar')
        plt.xticks(rotation=0)
        plt.title(f"Pollutant average by type for {indicator}")
    else:
        type_avg_df.plot(kind='bar')
        plt.xticks(rotation=0)
        plt.title("Pollutant average by type")

    plt.ylabel("Average value")
    plt.tight_layout()
    plt.show()
```

```python
type_avg('so2')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def location_avgs(state, indicator="so2"):
    # Keep only numeric columns for averaging
    numeric_cols = data[VALUE_COLS].select_dtypes(include='number').columns

    # Group by state and location, compute mean only for numeric cols
    locs = data[numeric_cols.tolist() + ['state', 'location']].groupby(['state', 'location']).mean(numeric_only=True)

    # Select the given state and reset index
    if state not in locs.index.get_level_values(0):
        print(f" State '{state}' not found in the data.")
        return
    
    state_avgs = locs.loc[state].reset_index()

    # Check if indicator exists in the dataframe
    if indicator not in state_avgs.columns:
        print(f" Indicator '{indicator}' not found in VALUE_COLS.")
        return

    # Plot the barplot
    plt.figure(figsize=(10, 6))
    sns.barplot(x='location', y=indicator, data=state_avgs, palette='viridis')
    plt.title(f"Location-wise average for {indicator.upper()} in {state}")
    plt.xlabel("Location")
    plt.ylabel(f"Average {indicator.upper()}")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
location_avgs("Bihar", "no2")
```
